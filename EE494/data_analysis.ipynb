```python
import pandas as pd
import numpy as np 
import os, glob
```

```python
list_of_files = glob.glob(os.path.join("/home/kepler42/EE494/EE494/Dataset_EE_lounge/", "EE494_data_*.csv"))
print(list_of_files)
```

```python
for f in list_of_files:
    print(f)
    df = pd.read_csv(f, sep=',',header=None)
    
    print(round(df.describe(),2))
```

```python
df_from_each_file = (round(pd.read_csv(f, sep=',',header=None).describe(),2) for f in list_of_files)
df_merged   = pd.concat(df_from_each_file, ignore_index=True)
df_merged.to_csv( "data_analysis.csv")
```
